```python
#!/usr/bin/env python
# coding: utf-8

# In[1]:


import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import warnings 
warnings.filterwarnings('ignore')


# In[2]:


df=pd.read_csv("/kaggle/input/walmart-dataset/Walmart.csv")
df


# # EDA 

# In[3]:


df.head()


# In[4]:


df.tail()


# In[5]:


df.shape


# In[6]:


df.size


# In[7]:


df.ndim


# In[8]:


df.describe()


# In[9]:


df.info()


# In[10]:


df.isnull().sum()


# # Encoding 

# In[24]:


from sklearn.preprocessing import LabelEncoder
a=LabelEncoder()
df['Date']=a.fit_transform(df['Date'])
df


# In[25]:


from sklearn.preprocessing import MinMaxScaler
c=MinMaxScaler(feature_range=(1,3))
d=['Date','Weekly_Sales','Temperature','Fuel_Price','CPI','Unemployment']
for i in d:
           df[i]=c.fit_transform(df[[i]])
df


# In[45]:


df.corr()


# In[44]:


sns.heatmap(df.corr(),annot=True,cmap='PiYG')


# In[37]:


sns.pairplot(df)


# In[41]:


sns.histplot(data=df,x='Weekly_Sales',color='r')
plt.title('Weekly sales CPi')
plt.xlabel('weekly')


# In[42]:


plt.bar(data=df,color='g',x=['Weekly_Sales','CPi'],height=20)
plt.title('Sales')


# In[43]:


df


# 

# In[ ]:





# In[ ]:





# In[ ]:





# In[ ]:





# In[ ]:





# In[ ]:





# In[ ]:





# In[ ]:





# In[ ]:





# In[ ]:





# In[ ]:
```
